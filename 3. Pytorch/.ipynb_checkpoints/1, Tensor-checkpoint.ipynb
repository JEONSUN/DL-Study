## 텐서(Tensor)

파이토치에서는 데이터의 기본 단위로 **텐서(tensor)**라는 것을 사용한다.


텐서는 다차원 배열(array)라고 정의할 수 있다. (numpy array와 역할이 비슷하다.)

> **벡터 -> 행렬 -> 텐서**



> #### **스칼라 : 차원이 없는 값**
> #### **벡터 : 1차원으로 구성된 값**
> #### **행렬 : 2차원으로 구성된 값**
> #### **텐서 : 3차원으로 구성된 값**
