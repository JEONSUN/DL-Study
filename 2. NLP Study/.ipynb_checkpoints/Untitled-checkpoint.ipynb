## 로지스틱 회귀(Logistic Regression)

풀고자하는 많은 문제 중에서는 두 개의 선택지 중에서 정답을 고르는 문제가 많다. 
ex) 예/아니오, 합격/불합격, 정상/스팸

이렇듯 둘 중하나를 결정하는 문제를 **이진 분류(Binary Classification)**라고 한다.

그리고 이런 문제를 풀기 위한 대표적인 알고리즘으로 **로지스틱 회귀(Logistic Regression)**라고 한다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 이진분류

성적과 합/불합 데이터를 생성해서 특정 점수를 얻었을때 합격, 불합격 여부를 판정하는 모델을 만들어보자


합격은 1, 불합격은 0으로 매핑시켰다.

```python
x = np.linspace(45,70,6)
y = [0,0,0,1,1,1]
```

```python
# 이진분류 그래프
plt.plot(x,y, '-o')
plt.grid()
plt.xlabel('score')
plt.ylabel('result')
plt.show()
```

해당 데이터를 그래프로 살펴보면 S자 형태로 나타난다는 걸 확인할 수 있다.

이러한 x와 y의 관계를 표현하기 위해서는 직선을 표현하는 함수(선형회귀)가 아니라 S자 형태로 표현할 수 있는 함수가 필요하다.

또한 실제값 y가 0또는 1이라는 두 가지 값뿐이기에 이 문제를 풀기 위해서는 예측값이 0과 1사이의 값을 가지도록하는 것이 보편적이다.

> 0과 1사이의 값을 확률로 해석하면 문제를 풀기가 훨씬 용이해지는데, **최종 예측값이 0.5보다 작으면 0으로 예측했다고 판단하고, 0.5보다 크면 1로 예측했다고 판단**하기 때문이다.
>
>그러나 선형회귀는 y값이 음의 무한대부터 양의 무한대와 같은 큰 수들도 가질 수 있는데 이는 분류 문제에 적합하지 않다고 판단하는 2번째 이유이다.


## 시그모이드 함수(Sigmoid Function)

#### Sigmoid 방정식
***
$$H(X) = {1\over1+e^{-(W_x+b)}}$$

***
선형 회귀와 마찬가지로 주어진 데이터에 가장 적합한 가중치 W와 편향 b를 구해야한다.

```python
# sigmoid 구현
def sigmoid(x) :
    return 1/(1+np.exp(-x))
x = np.arange(-5,5,0.1)
y = sigmoid(x)
```

```python
# W는 1, b는 0임을 가정
plt.plot(x,y,'b')
plt.plot([0,0],[1.0,0.0],':')
plt.title('Sigmoid Function')
plt.show()
```

시그모이드 함수는 출력값을 0과 1사이의 값으로 조정하여 반환하는데 x가 0일 때 0.5의 값을 가지며 x가 증가하면 1에 수렴한다.

```python
# w에 의한 변화 확인하기
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x,y1,'r',linestyle = '--', label = 'w = 0.5') # w의 값이 0.5일때
plt.plot(x,y2,'g', label = 'w = 1') # w의 값이 1일 떄
plt.plot(x,y3,'b',linestyle = '--', label = 'w = 2') # w의 값이 2일때
plt.plot([0,0],[1.0,0.0],':')
plt.title('Sigmoid Function')
plt.legend()
plt.show()
```

W의 값으로 인해 그래프의 경사도가 변하는 것을 확인할 수 있다. 선형회귀에서의 W는 직선의 기울기를 의미했지만, 여기서는 그래프의 경사도를 결정한다.

W의 값이 커지면 경사가 커지고, W의 값이 작아지면 경사가 작아진다.

```python
# b에 의한 변화 확인하기
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x,y1,'r',linestyle = '--', label = 'b = 0.5') # w의 값이 0.5일때
plt.plot(x,y2,'g', label = 'b = 1') # w의 값이 1일 떄
plt.plot(x,y3,'b',linestyle = '--', label = 'b = 2') # w의 값이 2일때
plt.plot([0,0],[1.0,0.0],':')
plt.title('Sigmoid Function')
plt.legend()
plt.show()
```

b의 값으로 인해 그래프가 이동하는 것을 확인할 수 있다.
