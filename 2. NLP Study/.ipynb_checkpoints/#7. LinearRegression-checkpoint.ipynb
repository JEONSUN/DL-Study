## Linear Regression

>시험 공부하는 시간을 늘리면 늘릴 수록 성적이 잘 나온다.<br>
>하루에 걷는 횟수를 늘릴 수록, 몸무게가 줄어든다.<br>
>집의 평수가 클수록, 집의 매매 가격은 비싼 경향이 있다

해당 문장들은 수학적으로 생각해보면 어떤 요인의 수치에 따라서 특정 요인의 수치가 영향을 받고 있다고 볼 수 있다.

조금 더 수학적인 표현을 써보면 어떤 변수의 값에 따라서 특정 변수의 값이 영향을 받고 있다고 볼 수 있다.

앞서 본 문장들처럼 다른 변수의 값을 변하게하는 변수를 **독립변수(x)**, 변수 x에 의해서 값이 종속적으로 변하는 변수를 **종속변수(y)**라고 정의하며, 선형 회귀는 한 개 이상의 독립 변수 x와 y의 선형 관계를 모델링한다. <br>
이때 독립 변수가 1개라면 **단순 선형 회귀**라고 한다.



### 단순 선형 회귀 (Simple Linear Regression)

### 수식
***
### $$y = Wx + b$$

***


여기서 독립 변수 x와 곱해지는 값 $W$를 머신 러닝에서는 **가중치(weight)**, 별도로 더해지는 값 $b$를 **편향(bias)**라고 부른다.

단순 선형 회귀에서는 가중치와 절편없이는 $y = x$ 하나의 직선밖에 표현을 못한다.

따라서 가중치와 절편 값을 적절히 찾아내면 x와 y의 관계를 적절히 모델링한 것이 된다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(1,10,20)
y1 = x1
y2 = (2*x1)+7
```

```python
# 가중치와 절편이 달라지면서 직선의 방정식도 달라진다
plt.plot(x1,y1, label = 'y = x')
plt.plot(x1,y2, label = 'y = 2x+7')
plt.plot(x1,x1+2*(np.log(x1+10))+7, label = 'y = x+2(log(x))+7')
plt.legend()
plt.show()
```

### 다중 선형 회귀(Multiple Linear Regression)

### 수식

***
### $$y = W_1x_1 + W_2x_2 + ...W_nx_n + b $$
***

좀 더 생각해보면 집의 평수가 클수록, 집의 방 개수, 역세권 유무와 같은 변수도 집의 매매가에 영향이 있다.

이러한 다수의 요소를 활용하여 1개의 반응변수 y에 x는 여러 개가 되었다. 이러한 경우를 **다중 선형 회귀**라고 부른다.


## 1. 자동 미분

```python
import tensorflow as tf
```

```python
w = tf.Variable(2.)
```

```python
# 임의로 2w**2 + 5라는 식을 생성
def f(w) :
    y = w**2
    z = 2*y + 5
    return z
```

```python
# w에 대한 미분값 저장
with tf.GradientTape() as tape :
    z = f(w)
    
gradients = tape.gradient(z, [w])
print(gradients)
```

## 2. 자동 미분을 이용한 선형 회귀 구현

```python
# 초기의 가중치 W와 b 선언(4,1 지정)
W = tf.Variable(4.0)
b = tf.Variable(1.0)
```

### hypothesis : 가설

- x와 y의 관계를 W와 b를 이용하여 식을 세우는 일
>적절한 직선의 방정식을 찾기위해선 적절한 W와 b를 찾아야한다.

```python
# 가설을 함수로서 지정
def hypothesis(x) :
    return W*x + b
```

```python
# 현재의 가설에서 임의의 입력값 입력
# 3.5*4 + 1 = 15
x_test = [3.5,5,5.5,6]
print(hypothesis(x_test).numpy())
```

### 비용 함수(Cost function) : 평균 제곱 오차(MSE)

W와 b를 찾기 위해서는 실제값과 가설로부터 얻은 예측값의 오차를 계산하는 식을 세우고, 이 식의 값을 최소화하는 최적의 W와 b를 찾아낸다

- 이 때 실제값과 예측값에 대한 오차에 대한 식을 **목적 함수(Objective function)** 또는 **비용 함수(Cost function)** 또는 **손실 함수(Loss function)**라고 한다.

비용 함수는 단순히 실제값과 예측값에 대한 오차를 표현하면 되는 것이 아니라, 예측값의 오차를 줄이는 일에 최적화된 식이어야한다.

```python
# 평균 제곱 오차를 손실 함수로 지정
def mse_loss(y_pred, y) :
    return tf.reduce_mean(tf.square(y_pred - y))
```

```python
# 데이터 생성
x = [1,2,3,4,5,6,7,8,9]
y = [11,22,33,44,53,66,77,87,95]
```

### 옵티마이저(Optimizer) : 경사하강법(Gradient Descent

- 비용 함수를 최소화하는 매개 변수인 W와 b을 찾기 위한 작업을 수행하는데, 이때 사용되는 알고리즘을 **옵티마이저(Optimizer)** 또는 **최적화 알고리즘**이라고 부른다.

이 옵티마이저를 통해 적절한 W와 b를 찾아내는 과정을 머신 러닝에서는 **학습(training)**이라고 부른다.

```python
# 경사하강법 사용, 학습률 0.01
optimizer = tf.optimizers.SGD(0.01)
```

```python
# 경사하강법 수행
for i in range(301) :
    with tf.GradientTape() as tape :
        # 현재 파라미터에 기반한 입력 x에 대한 예측값을 y_pred
        y_pred = hypothesis(x)
        
        # 평균 제곱 오차를 계산
        cost = mse_loss(y_pred, y)
        
    # 손실 함수에 대한 파라미터의 미분값 계산
    gradients = tape.gradient(cost, [W,b])
    
    # 파라미터 업데이트
    optimizer.apply_gradients(zip(gradients,[W,b]))
    
    if i % 10 == 0 :
        print("epoch : {:3} | W의 값 : {:5.4f} | b의 값 : {:5.4} | cost : {:5.6f}".format(i,W.numpy(),b.numpy(),cost))
```

W와 b값이 계속 업데이트 됨에 따라 cost가 지속적으로 줄어드는 것을 확인할 수 있따.

```python
# 학습된 w와 b의 값에 대해서 임의 입력을 넣었을 경우의 예측값을 확인해보자.
x_test = [3.5, 5, 5.5, 6]
print(hypothesis(x_test).numpy())
```

## 3. 선형회귀 케라스로 구현

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
x = [1,2,3,4,5,6,7,8,9] # 공부시간
y = [11,22,33,44,53,66,77,87,95] # 공부시간에 따른 성적

model = Sequential()

# input x와 output y의 차원은 모두 1, 선형 회귀이므로 activation = 'linear'
model.add(Dense(1, input_dim = 1, activation = 'linear'))

# 경사하강법 설정(학습률 0.01)
sgd = optimizers.SGD(lr = 0.01)

# 손실 함수(Loss function)은 평균제곱오차 mse를 사용
model.compile(optimizer = sgd, loss = 'mse', metrics = ['mse'])

# 주어진 x와 y데이터에 대해 오차를 최소화하는 작업을 300번 시도
model.fit(x,y, batch_size = 1, epochs = 300, shuffle = False)
```

```python
# 오차를 최소화하는 직선 그래프
plt.plot(x, model.predict(x), 'b',x,y,'k.')
plt.show()
```

```python
# 학습한 모델로 새로운 데이터 예측
print(model.predict([9.5]))
```

9시간 30분을 공부하면 약 98.6점을 얻는다고 예측함
